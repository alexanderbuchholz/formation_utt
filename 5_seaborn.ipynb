# Seaborn: des plots plus jolis

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

On va utiliser les données pima indians

```python
dataset = pd.read_csv('pima-indians-diabetes.csv', header=None)
dataset[[1,2,3,4,5]] = dataset[[1,2,3,4,5]].replace(0, np.NaN)
dataset.columns = ['num_pregnant', 'glucose', 'pressure', 'tricep', 'insulin', 'bmi', 'pedigree' , 'age', 'diab_class']
dataset = dataset.dropna()
```

```python
sns.distplot(dataset.glucose)
```

```python
sns.jointplot(dataset.age, dataset.num_pregnant) # ca vous étonne ? 
```

```python
sns.lmplot(x='bmi', y='pressure', data=dataset, hue='diab_class')
```

```python
sns.violinplot(x="diab_class", y="glucose", data=dataset)
plt.title('violin plot')
plt.xlabel('Diabetique ou pas') # on peut servir des fonctionalités matplotlib
```
