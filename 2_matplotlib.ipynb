```python
import numpy as np
# pour afficher les plots directement
%matplotlib inline 
import matplotlib.pyplot as plt
```

Un example assez simple pour vous familariser avec le langage: plotter une fonction sin et cosin

```python
x = np.linspace(0, 4*np.pi, 50)
sin_array = np.sin(x)
cos_array = np.cos(x)
plt.plot(x, cos_array)
plt.plot(x, sin_array)
```

```python
plt.plot(x, sin_array)
plt.show()
```

On ajoute des legende et d'autres fonctionalit

```python
plt.plot(x, cos_array, label="cos", linewidth=3, linestyle=":")
plt.plot(x, sin_array, label="sin", linewidth=3)
plt.legend()
plt.title('Deux courbes trigonometrique')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```

Des scatter-plots: 

```python
plt.scatter(x, cos_array, label="cos", linewidth=3, color="red")
plt.scatter(x, sin_array, label="sin", linewidth=3, color="green")
plt.legend()
plt.title('Deux courbes trigonometrique')
plt.xlabel("x")
plt.ylabel("y")
plt.xscale('symlog')
```

##Exercise matplotlib:

On va simuler des donnes artificielles pour deux groupes que l'on visualisera ensuite. 
Pour cela gnerer des observations 
$$y_i = a + b*x_i + e_i$$ 
et
$$z_j = c + d*v_j + \epsilon_i$$ 
o $i=1,...,50$, et 
$a=3.5$, $b=0.5$, $x_i \sim \mathcal{N}(1,1)$, $e_i \sim \mathcal{N}(0,3)$

o $j=1,...,40$, et 
$c=-2$, $d=0.7$, $v_j \sim \mathcal{N}(0.5,1)$, $\epsilon_j \sim \mathcal{N}(0,2)$


a) Simuler les deux jeux de donnes, fixer la seed du generateur alatoire a 42

b) Visualiser les jeux de donnes dans un plot joint

c) Visualiser les jeux de donnes dans deux subplot

d) Visualiser les jeux de donnes dans un plot joint avec deux couleures differentes pour les deux groupes

e) Ajouter une ligne de regression qui passe par les deux nuages

#Solution

```python
np.random.seed(42) # on fixe la racine du gnrateur alatoire
N1 = 50; a = 3.5; b = 0.5
N2 = 40; c = -2; d = 0.7
# le premier jeu de donnes
e = np.random.normal(loc=0, scale=3, size = N1)
x = np.random.normal(loc=1, scale=1, size = N1)
y = a+b*x+e

epsilon = np.random.normal(loc=0, scale=2, size = N2)
v = np.random.normal(loc=0.5, scale=1, size = N2)
z = c+d*v+epsilon

# une solution pour la question b)
plt.scatter(x,y, color="b")
plt.scatter(v,z, color="b")
plt.title("Le plot qui affiche tout en meme temps")
plt.show()

# on fait deux subplots
plt.figure()
plt.subplot(121)
plt.scatter(x,y)
plt.title('subplot 1')

plt.subplot(122)
plt.scatter(v,z)
plt.title('subplot 2')
plt.show()

# on plot pour les deux groupes
plt.scatter(x,y)
plt.scatter(v,z)
plt.title("Le plot qui affiche les groupes")
plt.show()

# on ajouter la ligne
y_fit = a+b*x
z_fit = c+d*v

plt.scatter(x,y)
plt.scatter(v,z)
plt.plot(x,y_fit)
plt.plot(v,z_fit)

plt.title("Le plot qui affiche les groupes")
plt.show()
```
