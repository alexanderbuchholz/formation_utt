# Intro matplotlib
matplotlib est une librairie puissante pour tous ce qui visualisation. Elle a plein d'option mais peut parfois etre lourde. Si vous avez le choix, optez pour une visualisation avec seaborn (plus tard). C'est assez jolie sans faire beaucoup de modification. Pour matplotlib voir

```python
import numpy as np
# pour afficher les plots directement
%matplotlib inline 
import matplotlib.pyplot as plt
```

Un example assez simple pour vous familariser avec le langage: plotter une fonction sin et cosin

```python
x = np.linspace(0, 4*np.pi, 50)
sin_array = np.sin(x)
cos_array = np.cos(x)
plt.plot(x, cos_array)
plt.plot(x, sin_array)
```

On ajoute des legende et d'autres fonctionalité

```python
plt.plot(x, cos_array, label="cos", linewidth=3)
plt.plot(x, sin_array, label="sin", linewidth=3)
plt.legend()
plt.title('Deux courbes trigonometrique')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```

Des scatter-plots: 

```python
plt.scatter(x, cos_array, label="cos", linewidth=3, color="red")
plt.scatter(x, sin_array, label="sin", linewidth=3, color="green")
plt.legend()
plt.title('Deux courbes trigonometrique')
plt.xlabel("x")
plt.ylabel("y")
plt.xscale('symlog')
```
