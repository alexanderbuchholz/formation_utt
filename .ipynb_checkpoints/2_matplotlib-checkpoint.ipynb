matplotlib est une librairie puissante pour tous ce qui visualisation. Elle a plein d'option mais peut parfois etre lourde. Si vous avez le choix, optez pour une visualisation avec seaborn (plus tard). C'est assez jolie sans faire beaucoup de modification. Pour matplotlib voir

```python
import numpy as np
# pour afficher les plots directement
%matplotlib inline 
import matplotlib.pyplot as plt
```

Un example assez simple pour vous familariser avec le langage: plotter une fonction sin et cosin

```python
x = np.linspace(0, 4*np.pi, 50)
sin_array = np.sin(x)
cos_array = np.cos(x)
plt.plot(x, cos_array)
plt.plot(x, sin_array)
```

On ajoute des legende et d'autres fonctionalité

```python
plt.plot(x, cos_array, label="cos", linewidth=3)
plt.plot(x, sin_array, label="sin", linewidth=3)
plt.legend()
plt.title('Deux courbes trigonometrique')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```

Des scatter-plots: 

```python
plt.scatter(x, cos_array, label="cos", linewidth=3, color="red")
plt.scatter(x, sin_array, label="sin", linewidth=3, color="green")
plt.legend()
plt.title('Deux courbes trigonometrique')
plt.xlabel("x")
plt.ylabel("y")
plt.xscale('symlog')
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data) # que font les arguements c et s ? 
plt.xlabel('entry a')
plt.ylabel('entry b')
```

Visualisation d'une gaussienne multivariée corrélée

```python
X = np.array([[0.4,1], [1,0.4]]).dot(np.random.normal(size=(2, 500)))
plt.scatter(X[0,:], X[1,:])
```

```python
plt.hist(X[0,:])
```

```python
plt.hist(X[1,:], bins=50)
```

Maintenant on travaille avec des entrées catégorielles

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
```

## Les subplots et enregistrer les fichiers

```python
plt.figure()
plt.subplot(211) # 2 ligne, 1 colone, element 1 dans la grille
plt.hist(X[0,:])
plt.subplot(212) # 2 ligne, 1 colone, element 2 dans la grille
plt.hist(X[1,:])
plt.savefig('deux_histogrammes_moches.pdf')
```
